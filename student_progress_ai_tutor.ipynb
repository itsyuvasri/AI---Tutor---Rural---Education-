```python
import matplotlib.pyplot as plt

# Simulated student progress data
sessions = [1, 2, 3, 4, 5]
student_scores = [20, 35, 55, 70, 85]

# Determine learning levels
levels = []
for score in student_scores:
    if score < 40:
        levels.append("Beginner")
    elif score < 70:
        levels.append("Intermediate")
    else:
        levels.append("Advanced")

# Print learning levels (for record)
print("Session-wise Learning Levels:")
for i in range(len(sessions)):
    print(f"Session {sessions[i]}: Score = {student_scores[i]}, Level = {levels[i]}")

# Plot graph
plt.figure()
plt.plot(sessions, student_scores, marker='o')
plt.xlabel("Learning Sessions")
plt.ylabel("Student Score")
plt.title("AI Tutor â€“ Student Learning Progress (SDG 4)")
plt.grid(True)
plt.show()
```
