```python

"""
===========================================================
Project Title : AI Tutor Chatbot for Personalized Rural Education
Author        : Yuvasri S
Platform      : Google Colab
Language      : Python
SDG Alignment : SDG 4 - Quality Education
===========================================================

Problem Statement:
Students in rural areas lack access to quality and personalized education
due to shortage of trained teachers, limited infrastructure, and lack of
individual attention.

Proposed Solution:
An AI-based chatbot tutor that interacts with students, asks questions,
evaluates responses, tracks performance, and adapts learning content based
on the studentâ€™s progress and comprehension level.

Technologies Used:
- Artificial Intelligence (rule-based decision making)
- Adaptive Learning Algorithms
- Basic Natural Language Processing (chat interaction)
- Learning Analytics
- Google Colab
- GitHub

===========================================================
"""

# ---------------------------------------------------------
# IMPORTS (basic â€“ works offline)
# ---------------------------------------------------------
import time

# ---------------------------------------------------------
# STUDENT PROFILE DATA STRUCTURE
# ---------------------------------------------------------
student_profile = {
    "name": "",
    "total_questions": 0,
    "correct_answers": 0,
    "current_level": "Easy",
    "history": []
}

# ---------------------------------------------------------
# QUESTION BANK (LEVEL-WISE)
# ---------------------------------------------------------
question_bank = {
    "Easy": [
        {"question": "What is 2 + 3?", "answer": "5"},
        {"question": "What is 10 - 6?", "answer": "4"},
        {"question": "What is 4 + 1?", "answer": "5"}
    ],
    "Medium": [
        {"question": "What is 6 Ã— 4?", "answer": "24"},
        {"question": "What is 20 Ã· 5?", "answer": "4"},
        {"question": "What is 15 + 10?", "answer": "25"}
    ],
    "Hard": [
        {"question": "What is 12 Ã— 8?", "answer": "96"},
        {"question": "What is 144 Ã· 12?", "answer": "12"},
        {"question": "What is 45 + 55?", "answer": "100"}
    ]
}

# ---------------------------------------------------------
# CHATBOT GREETING FUNCTION
# ---------------------------------------------------------
def greet_student():
    print("ðŸ¤– Welcome to AI Tutor Chatbot")
    print("ðŸ“˜ Personalized Learning for Rural Education\n")
    name = input("ðŸ‘‹ Please enter your name: ")
    student_profile["name"] = name
    print(f"\nHello {name}! Let's start your learning journey ðŸš€")
    time.sleep(1)

# ---------------------------------------------------------
# ASK QUESTION FUNCTION
# ---------------------------------------------------------
def ask_question():
    level = student_profile["current_level"]
    questions = question_bank[level]

    # Pick question based on attempts
    index = student_profile["total_questions"] % len(questions)
    current_q = questions[index]

    print(f"\nðŸ“š Level: {level}")
    print("ðŸ“ Question:", current_q["question"])

    user_answer = input("âœï¸ Your answer: ")

    student_profile["total_questions"] += 1

    if user_answer.strip() == current_q["answer"]:
        print("âœ… Correct answer! Well done ðŸ˜Š")
        student_profile["correct_answers"] += 1
        result = "Correct"
    else:
        print("âŒ Incorrect answer.")
        print("âœ” Correct answer is:", current_q["answer"])
        print("ðŸ“– Hint: Try solving step by step.")
        result = "Wrong"

    # Save history
    student_profile["history"].append({
        "question": current_q["question"],
        "your_answer": user_answer,
        "result": result,
        "level": level
    })

# ---------------------------------------------------------
# ADAPTIVE LEARNING LOGIC
# ---------------------------------------------------------
def adapt_learning_level():
    accuracy = student_profile["correct_answers"] / student_profile["total_questions"]

    if accuracy >= 0.75:
        if student_profile["current_level"] == "Easy":
            student_profile["current_level"] = "Medium"
            print("ðŸŽ¯ Level Up! Moving to MEDIUM level")
        elif student_profile["current_level"] == "Medium":
            student_profile["current_level"] = "Hard"
            print("ðŸ”¥ Excellent! Moving to HARD level")

    elif accuracy < 0.4:
        student_profile["current_level"] = "Easy"
        print("ðŸ” Revision needed. Returning to EASY level")

# ---------------------------------------------------------
# LEARNING ANALYTICS REPORT
# ---------------------------------------------------------
def show_progress_report():
    print("\nðŸ“Š LEARNING ANALYTICS REPORT")
    print("--------------------------------")
    print("ðŸ‘¤ Student Name      :", student_profile["name"])
    print("ðŸ“Œ Total Questions   :", student_profile["total_questions"])
    print("âœ… Correct Answers   :", student_profile["correct_answers"])
    print("ðŸ“š Current Level     :", student_profile["current_level"])

    accuracy = (student_profile["correct_answers"] /
                student_profile["total_questions"]) * 100
    print("ðŸ“ˆ Accuracy          :", round(accuracy, 2), "%")
    print("--------------------------------")

# ---------------------------------------------------------
# MAIN CHATBOT LOOP
# ---------------------------------------------------------
def start_chatbot():
    greet_student()

    while True:
        ask_question()
        adapt_learning_level()

        choice = input("\nðŸ‘‰ Do you want to continue learning? (yes/no): ").lower()

        if choice == "no":
            show_progress_report()
            print("\nðŸŽ‰ Thank you for learning with AI Tutor Chatbot!")
            print("ðŸ“˜ Keep learning. Keep growing.")
            break

# ---------------------------------------------------------
# PROGRAM STARTS HERE
# ---------------------------------------------------------
start_chatbot()
```
