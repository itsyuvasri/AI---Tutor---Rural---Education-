```python

from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt

# Actual and predicted levels
actual_levels = ["Beginner", "Beginner", "Intermediate", "Advanced", "Advanced"]
predicted_levels = ["Beginner", "Intermediate", "Intermediate", "Advanced", "Advanced"]

# Confusion Matrix
cm = confusion_matrix(
    actual_levels,
    predicted_levels,
    labels=["Beginner", "Intermediate", "Advanced"]
)

# Plot
disp = ConfusionMatrixDisplay(
    confusion_matrix=cm,
    display_labels=["Beginner", "Intermediate", "Advanced"]
)

disp.plot(cmap="Blues")
plt.title("Confusion Matrix â€“ AI Tutor Level Prediction")
plt.show()
```
